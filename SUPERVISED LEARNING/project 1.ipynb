```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
dataset=pd.read_csv('weight-height.csv')
```

```python
X_df_male = dataset.loc[dataset['Gender'] == 'Male']
X_df_female = dataset.loc[dataset['Gender'] == 'Female']
```

```python
Y_df_male=X_df_male['Weight']
Y_df_Female=X_df_female['Weight']
X_male=X_df_male.drop(['Gender','Weight'],axis=1)
X_female=X_df_female.drop(['Gender','Weight'],axis=1)
```

```python
X_male_train,X_male_test,Y_df_male_train,Y_df_male_test=train_test_split(X_male,Y_df_male,test_size=.2,random_state=42)
X_female_train,X_female_test,Y_df_Female_train,Y_df_Female_test=train_test_split(X_female,Y_df_Female,test_size=.2,random_state=42)
```

```python
sns.scatterplot(X_df_female)
```

```python
lr_m=LinearRegression()
lr_m.fit(X_male_train,Y_df_male_train)

Ypred_male_train=lr_m.predict(X_male_train)
Ypred_male_test=lr_m.predict(X_male_test)

lr_f=LinearRegression()
lr_f.fit(X_female_train,Y_df_Female_train)

Ypred_female_train=lr_m.predict(X_female_train)
Ypred_female_test=lr_m.predict(X_female_test)
```

```python
lr_m.score(X_male_test,Y_df_male_test)
```
